```python
#get variables without output
X=D.iloc[:,1:]
y=D.Alcohol
y = keras.utils.to_categorical(y-1)
```

```python
import sklearn
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=50,random_state=100)
```

```python
from sklearn import preprocessing
cr=preprocessing.StandardScaler()
X_train_cr=cr.fit_transform(X_train)
```

```python
X_train_cr
```

```python
y_test
```

```python
from keras.models import Sequential
from keras.layers import Dense,Flatten
#instanciation du modèle
model = Sequential()
#architecture
model.add(Dense(units=3,input_dim=13,activation="sigmoid"))
#compilation - algorithme d'apprentissage
model.compile(loss="categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
#apprentissage
model.fit(X_train_cr,y_train,epochs=150,batch_size=10)

```
